```python
# ---------------------------------------------------------
# SISTEMA BANCÁRIO POO - TITULAR: MARCOS VINÍCIUS
# ---------------------------------------------------------

class ContaBancaria:
    """Classe Pai que gerencia as operações básicas."""
    def __init__(self, titular, saldo_inicial=0):
        self.titular = titular
        self._saldo = saldo_inicial  # Atributo protegido

    def depositar(self, valor):
        if valor > 0:
            self._saldo += valor
            print(f"✅ Depósito de R${valor:.2f} realizado com sucesso!")
        else:
            print("❌ Valor de depósito inválido.")

    def sacar(self, valor):
        if 0 < valor <= self._saldo:
            self._saldo -= valor
            print(f"✅ Saque de R${valor:.2f} realizado.")
        else:
            print(f"❌ Saque negado. Saldo insuficiente ou valor inválido.")

    def exibir_info(self):
        return f"Titular: {self.titular} | Saldo Atual: R${self._saldo:.2f}"


class ContaPoupanca(ContaBancaria):
    """Classe Filha que herda de ContaBancaria e adiciona juros."""
    def __init__(self, titular, saldo_inicial=0, taxa_juros=0.005):
        # super() inicializa a classe pai
        super().__init__(titular, saldo_inicial)
        self.taxa_juros = taxa_juros

    def render_juros(self):
        valor_rendimento = self._saldo * self.taxa_juros
        self._saldo += valor_rendimento
        print(f"💰 Rendimento aplicado: R${valor_rendimento:.2f} (Taxa: {self.taxa_juros*100}%)")

# --- TESTANDO O CÓDIGO ---

# 1. Criando a conta conforme solicitado
conta_marcos = ContaPoupanca(titular="Marcos Vinícius", saldo_inicial=2000)

print("-" * 40)
print(conta_marcos.exibir_info())

# 2. Realizando operações
conta_marcos.depositar(500)      # Novo saldo: 2500
conta_marcos.render_juros()      # Rendimento sobre 2500
conta_marcos.sacar(300)          # Saque final

# 3. Resultado Final
print("-" * 40)
print("STATUS FINAL DA CONTA:")
print(conta_marcos.exibir_info())
print("-" * 40)
```
